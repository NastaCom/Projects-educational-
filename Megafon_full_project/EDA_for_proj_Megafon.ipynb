### **Рассмотрим активность подключения услуг в течение 6 месяцев**

```python
plt.figure(figsize=(20,10))
plt.hist(df_train.loc[(df_train['target']==0, 'date') ], bins=50, rwidth=5, stacked=False)
plt.hist(df_train.loc[(df_train['target']==1, 'date') ], bins=50, rwidth=5, stacked=False, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize = (15,15))
for i in [1,2,4,5,6,7,8,9]:
    plt.plot(df_train.loc[(df_train['vas_id']==i) & (df_train['target']==1), 'date'].value_counts().sort_index(),
            label = f'Услуга {i}')
    plt.title('Количество подключений услуги')
    plt.legend()
#     plt.yscale('log')
    plt.xticks(rotation=45);
```

```python
plt.figure(figsize = (15,15))
for i in [1,2,4,5,6,7,8,9]:
    plt.plot(df_train.loc[(df_train['vas_id']==i) & (df_train['target']==0), 'date'].value_counts().sort_index(),
            label = f'Услуга {i}')
    plt.title('Количество отказов от услуги')
    plt.legend()
    plt.xticks(rotation=45);
```

```python
plt.figure(figsize = (15,15))
for i in [1,2,4,5,6,7,8,9]:
    plt.plot(df_train.loc[(df_train['vas_id']==i), 'date'].value_counts().sort_index(),
            label = f'Услуга {i}')
    plt.title('Количество предложений услуги')
    plt.legend()
    plt.xticks(rotation=45);
```

```python
plt.figure(figsize = (15,15))
for i in [1,2,4,5,6,7,8,9]:
    plt.plot(df_train.loc[(df_train['vas_id']==i) & (df_train['target']==1), 'date'].value_counts().sort_index() / 
             df_train.loc[(df_train['vas_id']==i), 'date'].value_counts().sort_index(),
            label = f'Услуга {i}')
    plt.title('Отношение количества подключений услуги к количеству предложений')
    plt.legend()
    plt.xticks(rotation=45);
```

```python
df = (df_train[df_train['target'] == 1].groupby(['vas_id', 'date']).count()['id'] / 
      df_train[df_train['target'] == 0].groupby(['vas_id', 'date']).count()['id']).reset_index()

df.rename(columns={'id':'agree_count/reject_count'}, inplace=True)

plt.figure(figsize = (12,12))
for i in [1,2,4,5,6,7,8,9]:
    plt.plot(df[df['vas_id']==i]['date'], df[df['vas_id']==i]['agree_count/reject_count'], label = f'Услуга {i}')
    plt.title('Отношение количества подключений услуги к количеству отказов')
    plt.legend()
    plt.xticks(rotation=90);
```

По графикам наблюдается всплеск подключения услуг в середине и конце ноября. Возможно на всплеск повлияла акция...

```python
df.groupby(['date', 'vas_id']).max().sort_values(by='agree_count/reject_count', ascending=False).head()
```
