 5.2 Больше лучших клиентов, чем потерянных клиентов?
 
 Т.е. выяснить больше ли клиентов, которые покупали совсем недавно, чаще всего и тратят больше всего, 
 чем клиентов, у которых последняя покупка была давно, и покупали мало и тратили мало.

```python
date_last = df['Дата_изм'].max()
date_last
```

```python
client_diff = df.groupby(['Клиент', 'Транзакция', 'Дата_изм'])['Количество', 'Сумма'].sum() \
                .reset_index().drop_duplicates() \
                .groupby('Клиент', as_index = False).agg({'Дата_изм': lambda x: (date_last - x.max()).days,
                                       'Транзакция': lambda x: len(x),
                                       'Сумма': lambda x: x.sum()})

client_diff
```

```python
best_client = client_diff[(client_diff['Транзакция']>=30) & 
                          (client_diff['Сумма']>=400000) &
                          (client_diff['Дата_изм']<100)]
best_client.shape
```

```python
best_client
```

```python
bad_client = client_diff[(client_diff['Транзакция']<2) & 
                          (client_diff['Сумма']<1000)&
                          (client_diff['Дата_изм']>100)]

bad_client.shape
```

```python
bad_client
```

```python
best_client_row = best_client['Клиент']
bad_client_row = bad_client['Клиент']

best_client_count = best_client['Клиент'].count()
bad_client_count = bad_client['Клиент'].count()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
best_client, bad_client = ax.flatten()

best_client.hist(best_client_row, best_client_count)
best_client.set_title('Хорошие клиенты')

bad_client.hist(bad_client_row, bad_client_count)
bad_client.set_title('Плохие клиенты')


fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

##### Можно сделать вывод о том, что потерянных клиентов намного больше чем лучших (Что в свою очередь логично. Обычно разово покупающих клиентов больше, чем постоянных)

### 6. Кластеризация по товарам

```python
#Ваш код
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans = kmeans.fit(df[['Количество', 'Сумма']])
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Метод локтя')
plt.xlabel('Номер кластера')
plt.ylabel('Внутрикластерная сумма квадратов')
plt.show()
```

##### Заметное преломление вижу на двух и на трех кластерах. Возьму 3 кластера

```python
df_clean = df[['Количество', 'Товар', 'Сумма']]
df_clean
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 0).fit(df_clean[['Количество', 'Товар']])
predict = kmeans.predict(df_clean[['Количество', 'Товар']])
df_clean['Кластер'] = predict
df_clean.head()
```

```python
sns.boxplot(x=df_clean['Кластер'], y=df_clean['Товар'], data=df_clean)
```

### 7. Важные признаки

```python
#Ваш код
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = df[['Транзакция', 'Клиент', 'Товар', 'Количество', 'Сумма']]
y = df.iloc[:, -1]
y = y.astype('int')

bestfeatures = SelectKBest(score_func = chi2, k = 'all')
fit = bestfeatures.fit(X, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

featureScores = pd.concat([dfcolumns, dfscores], axis = 1)
featureScores.columns = ['Specs', 'Score']  
print(featureScores.nlargest(10, 'Score'))  
```

##### Самый важный признак в наших данных это Сумма, а самый бесполезный - Количество

### 8.	Рекомендации

##### Для системы рекомендаций создам удобную таблицу

```python
data = pd.melt(df.set_index('Клиент')['Товар'].apply(pd.Series).reset_index(), 
             id_vars=['Клиент'],
             value_name='Товар') \
    .dropna().drop(['variable'], axis=1) \
    .groupby(['Клиент', 'Товар']) \
    .agg({'Товар': 'count'}) \
    .rename(columns={'Товар': 'purchase_count'}) \
    .reset_index() \
    .rename(columns={'Товар': 'ТоварId'})
data['ТоварId'] = data['ТоварId'].astype(np.int64)
```

```python
print(data.shape)
data.head()
```

##### Преобразуем нашу таблицу в матрицу Клиент-Товар

```python
col = ['Клиент','ТоварId','purchase_count']

w1 = data[col]
w1 = w1.dropna(axis=0)
w1 = w1.drop_duplicates(['Клиент','ТоварId'])

w1_pivot = w1.pivot(index= 'Клиент',columns='ТоварId',values='purchase_count').fillna(0)
w1_pivot_matrix = csr_matrix(w1_pivot)
```
